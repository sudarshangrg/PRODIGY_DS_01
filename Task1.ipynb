### Importing necessary python libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population.csv")
```

```python
data
```

```python
data.head(5)
```

```python
data.describe
```

```python
#total number of unique countries
print(data['Country Name'].unique())
print('\n the total number of countries:',data['Country Name'].nunique())
```

```python
#List of countries for the graph
countries_to_plot = ['Nepal', 'Zambia', 'Brazil', 'Japan', 'Indonesia', 'France']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]
```

```python
filtered_data
```

```python
#bar chart for the 5 selected countries in 2022
plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
